```python
import numpy as np
import skfuzzy as fuzz
import skfuzzy.membership as mf
import matplotlib.pyplot as plt
x_hw = np.arange(0, 101, 1)
x_physics = np.arange(0, 101, 1)
x_maths = np.arange(0, 101, 1)
x_chemistry = np.arange(0, 101, 1)
y_risk = np.arange(0,100,1)
print("\n\n-------------------------------------------------\n\n")
print("Enter your data...\n\n.")
input_hw = int(input("hard work (0-100):"))
input_physics = int(input("physics marks (0-100):"))
input_maths = int(input("maths (0-100):"))
input_chemistry = int(input("chemistry (0-100): "))
print("\n\nThank you for the values you entered.")
print("\n\nCalculations are being made. Please wait...\n\n")
hw_low = mf.trapmf(x_hw, [-30, -5, 30, 40])
hw_mid = mf.trapmf(x_hw, [30, 40, 50, 60])
hw_high = mf.trapmf(x_hw, [50, 60, 100, 100])
physics_low = mf.trapmf(x_physics, [-30, -5, 30, 40])
physics_mid = mf.trapmf(x_physics, [30, 40, 50, 60])
physics_high = mf.trapmf(x_physics, [50, 60, 100, 100])
maths_low = mf.trapmf(x_maths, [-30, -5, 30, 40])
maths_mid = mf.trapmf(x_maths, [30, 40, 50, 60])
maths_high = mf.trapmf(x_maths, [50, 60, 100, 100])
chemistry_low = mf.trapmf(x_chemistry, [-30, -5, 30, 40])
chemistry_mid = mf.trapmf(x_chemistry, [30, 40, 50, 60])
chemistry_high = mf.trapmf(x_chemistry, [50, 60, 100, 100])
risk_not = mf.trapmf(y_risk, [0,0,15,20]) #[0 ,0 ,5 ,10])
risk_little = mf.trapmf(y_risk, [15,20,35,40]) #[5 ,10 ,15 ,20])
risk_mid = mf.trapmf(y_risk, [35,40,55,60]) #[15 ,20 ,25 ,30])
risk_high = mf.trapmf(y_risk, [55,60,75,80]) #[25 ,30 ,35 ,40])
risk_veryHigh = mf.trapmf(y_risk, [75,80,95,100]) #[35, 40, 100, 100])
fig, (ax0, ax1, ax2, ax3, ax4) = plt.subplots(nrows = 5, figsize =(10, 25))
ax0.plot(x_hw, hw_low, 'r', linewidth = 2, label = 'low')
ax0.plot(x_hw, hw_mid, 'g', linewidth = 2, label = 'mid')
ax0.plot(x_hw, hw_high, 'b', linewidth = 2, label = 'high')
ax0.set_title('hardwork')
ax0.legend()
ax1.plot(x_physics, physics_low, 'r', linewidth = 2, label = 'Low')
ax1.plot(x_physics, physics_mid, 'g', linewidth = 2, label = 'Middle')
ax1.plot(x_physics, physics_high, 'b', linewidth = 2, label = 'High')
ax1.set_title('Physics')
ax1.legend()
ax2.plot(x_maths, maths_low, 'r', linewidth = 2, label = 'Low')
ax2.plot(x_maths, maths_mid, 'g', linewidth = 2, label = 'Middle')
ax2.plot(x_maths, maths_high, 'b', linewidth = 2, label = 'High')
ax2.set_title('Maths')
ax2.legend()
ax3.plot(x_chemistry, chemistry_low, 'r', linewidth = 2, label = 'Low')
ax3.plot(x_chemistry, chemistry_mid, 'g', linewidth = 2, label = 'Middle')
ax3.plot(x_chemistry, chemistry_high, 'b', linewidth = 2, label = 'High')
ax3.set_title('Chemistry')
ax3.legend()
ax4.plot(y_risk, risk_not, 'r', linewidth = 2, label = 'Not')
ax4.plot(y_risk, risk_little, 'g', linewidth = 2, label = 'Little')
ax4.plot(y_risk, risk_mid, 'b', linewidth = 2, label = 'Middle')
ax4.plot(y_risk, risk_high, 'y', linewidth = 2, label = 'High')
ax4.plot(y_risk, risk_veryHigh, 'm', linewidth = 2, label = 'Very High')
ax4.set_title('Risk')
ax4.legend()
plt.tight_layout()
hw_fit_low = fuzz.interp_membership(x_hw, hw_low, input_hw)
hw_fit_mid = fuzz.interp_membership(x_hw, hw_mid, input_hw)
hw_fit_high = fuzz.interp_membership(x_hw, hw_high, input_hw)
physics_fit_low = fuzz.interp_membership(x_physics, physics_low, input_physics)
physics_fit_mid = fuzz.interp_membership(x_physics, physics_mid, input_physics)
physics_fit_high = fuzz.interp_membership(x_physics,physics_high ,input_physics)
maths_fit_low = fuzz.interp_membership(x_maths, maths_low, input_maths)
maths_fit_mid = fuzz.interp_membership(x_maths, maths_mid, input_maths)
maths_fit_high = fuzz.interp_membership(x_maths, maths_high, input_maths)
chemistry_fit_low = fuzz.interp_membership(x_chemistry,chemistry_low,input_chemistry)
chemistry_fit_mid = fuzz.interp_membership(x_chemistry,chemistry_mid,input_chemistry)
chemistry_fit_high = fuzz.interp_membership(x_chemistry,chemistry_high,input_chemistry)
rule1 = np.fmin(np.fmin(np.fmin(np.fmin(physics_fit_high,maths_fit_high),chemistry_fit_high), hw_fit_high),risk_not)
rule2 = np.fmin(np.fmin(np.fmin(np.fmin(physics_fit_high,maths_fit_high),chemistry_fit_high), hw_fit_mid),risk_little)
rule3 = np.fmin(np.fmin(np.fmin(np.fmin(physics_fit_high,maths_fit_high),chemistry_fit_high), hw_fit_low),risk_mid)
rule4 = np.fmin(np.fmin(np.fmin(np.fmin(physics_fit_mid,maths_fit_mid),chemistry_fit_mid), hw_fit_mid),risk_high)
rule5 = np.fmin(np.fmin(np.fmin(np.fmin(physics_fit_mid,maths_fit_mid),chemistry_fit_mid), hw_fit_low),risk_veryHigh)
rule6 = np.fmin(np.fmin(np.fmin(np.fmin(physics_fit_low,maths_fit_low),chemistry_fit_low), hw_fit_high),risk_little)
rule7 = np.fmin(np.fmin(np.fmin(np.fmin(physics_fit_low,maths_fit_low),chemistry_fit_low), hw_fit_mid),risk_high)
rule8 = np.fmin(np.fmin(np.fmin(np.fmin(physics_fit_low,maths_fit_low),chemistry_fit_low), hw_fit_low),risk_veryHigh)
rule5 = np.fmin(np.fmin(np.fmin(np.fmin(physics_fit_mid,maths_fit_mid),chemistry_fit_mid), hw_fit_high),risk_not)
out_not = np.fmax(rule1,rule6)
out_little = np.fmax(rule2,rule2)
out_mid = np.fmax(rule3,rule3)
out_high = np.fmax(rule4, rule7)
out_veryHigh = np.fmax(rule5,rule8)
risk0 = np.zeros_like(y_risk)
fig, ax0 = plt.subplots(figsize = (7, 4))
ax0.fill_between(y_risk, risk0, out_not, facecolor = 'r', alpha = 0.7)
ax0.plot(y_risk, risk_not, 'r', linestyle = '--')
ax0.fill_between(y_risk, risk0, out_little, facecolor = 'g', alpha = 0.7)
ax0.plot(y_risk, risk_little, 'g', linestyle = '--')
ax0.fill_between(y_risk, risk0, out_mid, facecolor = 'b', alpha = 0.7)
ax0.plot(y_risk, risk_mid, 'b', linestyle = '--')
ax0.fill_between(y_risk, risk0, out_high, facecolor = 'y', alpha = 0.7)
ax0.plot(y_risk, risk_high, 'y', linestyle = '--')
ax0.fill_between(y_risk, risk0, out_veryHigh, facecolor = 'm', alpha = 0.7)
ax0.plot(y_risk, risk_veryHigh, 'm', linestyle = '--')
ax0.set_title('Out of the Risk')
plt.tight_layout()
out_risk = np.fmax(np.fmax(np.fmax(np.fmax(out_not, out_little),out_mid),out_high), out_veryHigh)
defuzzified = fuzz.defuzz(y_risk, out_risk, 'centroid')
result = fuzz.interp_membership(y_risk, out_risk, defuzzified)
print("Outcome of risk is:", defuzzified)
fig, ax0 = plt.subplots(figsize=(7, 4))
ax0.plot(y_risk, risk_not, 'r', linewidth = 0.5, linestyle = '--')
ax0.plot(y_risk, risk_little, 'g', linewidth = 0.5, linestyle = '--')
ax0.plot(y_risk, risk_mid, 'b', linewidth = 0.5, linestyle = '--')
ax0.plot(y_risk, risk_high, 'y', linewidth = 0.5, linestyle = '--')
ax0.plot(y_risk, risk_veryHigh, 'm', linewidth = 0.5, linestyle = '--')
ax0.fill_between(y_risk, risk0, out_risk, facecolor = 'Orange', alpha = 0.7)
ax0.plot([defuzzified , defuzzified], [0, result], 'k', linewidth = 1.5, alpha=0.9)
ax0.set_title('Centroid Deffuzification')
plt.tight_layout()
```
